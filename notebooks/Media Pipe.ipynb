````python
import os
import pandas as pd
````

# Organizing Dataset

````python
paths = []
for sign in os.listdir('/content/drive/MyDrive/SIGN/Dataset'):
  videos = os.listdir(f'/content/drive/MyDrive/SIGN/Dataset/{sign}')
  for video in range(1,len(videos)+1):
    if video == len(videos):
      os.rename(f'/content/drive/MyDrive/SIGN/Dataset/{sign}/{videos[video-1]}', f'/content/drive/MyDrive/SIGN/Dataset/{sign}/{sign}_{video}test.MP4')
      paths.append(f'/content/drive/MyDrive/SIGN/Dataset/{sign}/{sign}_{video}test.MP4')
    else:
      os.rename(f'/content/drive/MyDrive/SIGN/Dataset/{sign}/{videos[video-1]}', f'/content/drive/MyDrive/SIGN/Dataset/{sign}/{sign}_{video}.MP4')
      paths.append(f'/content/drive/MyDrive/SIGN/Dataset/{sign}/{sign}_{video}.MP4')
````

````python
video_details = pd.DataFrame(columns = ['Video Path', 'Video Name', 'Label', 'Type'])

for path in paths:
  parts = path.split('/')
  label, _ = parts[-1].split('_')
  video_type = 'test' if 'test' in path else 'train'
  video_details.loc[-1] = [path, parts[-1], label, video_type]
  video_details.index = video_details.index + 1
  video_details = video_details.sort_index()

video_details.to_csv('/content/video_details.csv')
````
