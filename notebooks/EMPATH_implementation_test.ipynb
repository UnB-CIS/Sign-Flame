# Organizing Dataset

````python
paths = []
traintest = []
for _traintest in os.listdir('/content/drive/MyDrive/SIGN/Dataset_processed'):
  traintest.append(_traintest)

traintest = traintest[:-1]

for traintest_ in traintest:
  for sign in os.listdir(f'/content/drive/MyDrive/SIGN/Dataset_processed/{traintest_}'):
    videos = os.listdir(f'/content/drive/MyDrive/SIGN/Dataset_processed/{traintest_}/{sign}')
    for video in videos:
      paths.append(f'/content/drive/MyDrive/SIGN/Dataset_processed/{traintest_}/{sign}/{video}')
````

# Train Test

````python
video_details = pd.DataFrame(columns = ['Video Path', 'Video Name', 'Label', 'Type'])

for path in paths:
  parts = path.split('/')
  label = parts[-2]
  video_type = 'test' if 'test' in path else 'train'
  video_details.loc[-1] = [path, parts[-1], label, video_type]
  video_details.index = video_details.index + 1
  video_details = video_details.sort_index()

video_details.to_csv('/content/drive/MyDrive/SIGN/Documentos/video_details.csv')
````
